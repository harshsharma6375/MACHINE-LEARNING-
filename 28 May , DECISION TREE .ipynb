```python
import numpy as np 
import pandas as pd 
```

```python
df = pd.read_csv("C:\\Users\\HP\\OneDrive\\Desktop\\DATASET\\insurance - insurance.csv")
```

## 

```python
df.head(2)
```

```python
df = pd.get_dummies(df,columns= ['sex' , 'smoker' , 'region'])
```

```python
df = df.astype(int)
```

```python
x = df.drop(columns = ['charges'])
y = df['charges']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test , y_train , y_test = train_test_split(x,y,test_size = 0.2 , random_state=42)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dt = DecisionTreeRegressor()
```

```python
dt.fit(x_train , y_train)
```

```python
y_pred = dt.predict(x_test)
```

```python
from sklearn.metrics import r2_score  ## we use r2_score for regression in numerical based data 
```

```python
r2_score(y_test , y_pred)
```

# DESICION  TREE  BASED ON CLASSIFIER

A Decision Tree Classifier splits data based on feature values to classify samples into distinct categories. It's like a flowchart where each internal node represents a decision based on a feature, and each leaf node represents a class label.



```python
df = pd.read_csv("C:\\Users\\HP\\OneDrive\\Desktop\\DATASET\\Social_Network_Ads - Social_Network_Ads.csv")
```

```python
df.head(2)
```

```python
df = pd.get_dummies(df , columns = ['Gender'])
```

## 

```python
df.head(2)
```

```python
x  = df.drop(columns = ['Purchased'])
y = df['Purchased']
```

```python
x_train,x_test , y_train , y_test = train_test_split(x,y,test_size = 0.2 , random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier()
```

```python
dt.fit(x_train , y_train)
```

```python
y_pred = dt.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test , y_pred)
```
